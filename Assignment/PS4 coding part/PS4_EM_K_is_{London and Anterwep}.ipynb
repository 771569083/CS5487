```python
import pandas as pd
import numpy as np
import random
import scipy
from scipy.special import factorial
```

```python
# build or load the dataset.

London_data = np.concatenate ( (np.zeros((229)), np.ones((211)), 2 * np.ones((93)), 
3 * np.ones((35)), 4 * np.ones((7)), 5 * np.ones((1)) ) )
London_data

Antwep_data = np.concatenate ( (np.zeros((325)), np.ones((115)), 2 * np.ones((67)), 
3 * np.ones((30)), 4 * np.ones((18)), 5 * np.ones((21)) ) )

sample_data = np.concatenate((London_data, Antwep_data))

np.random.shuffle(sample_data)
# print(sample_data)

# len(sample_data)
sample_data
```

```python
mean_london = (211 + 2*93 + 3*35 + 4*7 + 5)/576

mean_anterwep = (115 + 2*67 + 3*30 + 4*18 + 5*21)
```

```python
from math import exp

def Single_Poisson_Distribution(sample, lbda, Phi):
    result = (exp(-lbda) * np.power(lbda, sample) )/factorial(sample)
    return result

def E_step(dataset, lbda_london, Phi_london, lbda_anterwep, Phi_anterwep):

    # alpha is each cluster's Poisson Distrbution coeffcient
    # responsibility is each cluster's responsibility or the Z_ij in Lecture.
    # P_D_con_thetais the P(x=ki|thetaj)

    Likelihood_multi_Prior_london = Phi_london*Single_Poisson_Distribution(dataset, lbda_london, Phi_london)

    Likelihood_multi_Prior_anterwep = Phi_anterwep*Single_Poisson_Distribution(dataset, lbda_anterwep, Phi_anterwep)

    P_D_con_theta = Likelihood_multi_Prior_london + Likelihood_multi_Prior_anterwep

    respon_london = Likelihood_multi_Prior_london/P_D_con_theta

    respon_anterwep = Likelihood_multi_Prior_anterwep/P_D_con_theta
    return respon_london, respon_anterwep

def M_step(dataset, respon_london, respon_anterwep):
    # according to the PS4.3 a question anwser
    Phi_london = np.sum(respon_london)/len(dataset)
    lbda_london = np.dot(respon_london, dataset)/np.sum(respon_london)

    Phi_anterwep = np.sum(respon_anterwep)/len(dataset)
    lbda_anterwep = np.dot(respon_anterwep, dataset)/np.sum(respon_anterwep)

    return Phi_london, lbda_london, Phi_anterwep, lbda_anterwep

def EM_train(dataset, iterations=1000):
    lbda_london = mean_london
    Phi_london = 0.5
    lbda_anterwep = mean_anterwep
    Phi_anterwep = 0.5

    step = 0
    while (step <= iterations):
        step += 1
        
        respon_london, respon_anterwep = E_step(dataset, lbda_london, Phi_london, lbda_anterwep, Phi_anterwep)
        
        print('In step{}, respon_london is {} and respon_anterwep is {}'.format(step, respon_london, respon_anterwep) )
        print('=======================================================================')

        Phi_london, lbda_london, Phi_anterwep, lbda_anterwep = M_step(dataset, respon_london, respon_anterwep)
    
    return Phi_london, lbda_london, Phi_anterwep, lbda_anterwep
```

```python
Phi_london, lbda_london, Phi_anterwep, lbda_anterwep = EM_train(sample_data)
```
